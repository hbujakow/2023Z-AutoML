```python
import pandas as pd
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import RFE, RFECV
import numpy as np
from sklearn.metrics import f1_score, balanced_accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer

from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import scipy
import matplotlib.pyplot as plt
```

Odczytanie danych

```python
train_data = pd.read_csv('/content/drive/MyDrive/AutoML/HW2/artificial_train.data', sep=' ', header = None).drop(columns=[500])
train_data.columns = ['col_'+f'{i+1}' for i in range(500)]
test_data = pd.read_csv('/content/drive/MyDrive/AutoML/HW2/artificial_test.data', sep=' ', header = None).drop(columns=[500])
test_data.columns = ['col_'+f'{i+1}' for i in range(500)]
```

Podział danych na treningowe i walidacyjne

```python
x_train, x_val, y_train, y_val = train_test_split(train_data, labels, test_size = 0.3)
```

### Selekcja kolumn

```python
etc = ExtraTreesClassifier()
rfe = RFE(etc)
rfe.fit(x_train, y_train.to_numpy().ravel())
cols_to_drop = x_train.columns.values[np.logical_not(rfe.support_)]
x_train_dropped = x_train.drop(columns = cols_to_drop)

rfecv = RFECV(etc, cv=3)
rfecv.fit(x_train_dropped, y_train.to_numpy().ravel())
cols_to_drop2 = x_train_dropped.columns.values[np.logical_not(rfecv.support_)]
x_train_final = x_train_dropped.drop(columns = cols_to_drop2)

x_val_final = x_val.drop(columns=cols_to_drop).drop(columns=cols_to_drop2)
```

```python
x_test = test_data.drop(columns=cols_to_drop).drop(columns=cols_to_drop2)
```

### Skalowanie danych

```python
sc = StandardScaler()
train_df = pd.DataFrame(sc.fit_transform(x_train_final),columns = x_train_final.columns) #sc.fit_transform(x_train_final)
train_df['class'] =  y_train.to_numpy().ravel()
val_df = pd.DataFrame(sc.transform(x_val_final),columns = x_train_final.columns)
val_df['class'] =  y_val.to_numpy().ravel()
test_df = pd.DataFrame(sc.transform(x_test),columns = x_test.columns)
```

### Zapisanie przerobionych danych na dysk

```python
train_df.to_csv('/content/drive/MyDrive/AutoML/HW2/preprocessed_train.data', sep=' ')
val_df.to_csv('/content/drive/MyDrive/AutoML/HW2/preprocessed_val.data', sep=' ')
test_df.to_csv('/content/drive/MyDrive/AutoML/HW2/preprocessed_test.data', sep = ' ')
```

## Auto Sklearn Classifier

```python
from autosklearn.classification import AutoSklearnClassifier
from autosklearn.experimental.askl2 import AutoSklearn2Classifier
from autosklearn.metrics import roc_auc,balanced_accuracy

import sklearn.datasets
import sklearn.metrics
import sklearn.model_selection
```

```python
settings = {
  "time_left_for_this_task": 600,  # seconds
  "seed": 42,
  "metric": balanced_accuracy,
  "n_jobs": 4,
}

# This will only be used by autosklearn 1 while autosklearn 2 will automatically
# select a strategy
resampling_strategy = "holdout"
```

```python
askl1 = AutoSklearnClassifier(
    **settings,
    resampling_strategy=resampling_strategy
)
askl1.fit(x_train, y_train)
```

```python
from pprint import pprint

# print(f"Auto-sklearn 1.0 | train = {train_scoring_function(askl1)} | test = {scoring_function(askl1)}")
print(f"Selected `resampling-strategy` = {askl1.resampling_strategy}")
print(f"Selected `resampling-strategy-arguments` = {askl1.resampling_strategy_arguments}")
```

```python
# Some quick summary statistics
print(askl1.sprint_statistics())
```

```python
# The leaderboard shows all the models during the optimization process,
# see this link for arguments if you want to see more!
# https://automl.github.io/auto-sklearn/master/api.html#autosklearn.classification.AutoSklearnClassifier.leaderboard
leaderboard = askl1.leaderboard(sort_by="model_id", ensemble_only=True)
print(leaderboard)
```

```python
# Show all the models in the final produced ensemble
print(askl1.show_models())
```

```python
predictions = askl1.predict(x_val)
```
